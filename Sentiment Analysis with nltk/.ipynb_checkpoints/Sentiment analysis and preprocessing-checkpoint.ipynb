```python
# Import libarry :
import numpy as np
import pandas as pd 
```

```python
import nltk
#nltk.download()
```

```python
# Import data :
df = pd.read_csv(r"D:\Dataset files_ML\sentimentdataset.csv")
```

```python
df.head()
```

```python
# Import Tokenizer :
from nltk.tokenize import word_tokenize
```

```python
# Tokenize each text entry :
text_tokens = [word_tokenize(i) for i in df["Text"]]
```

```python
text_tokens
```

```python
from nltk.tokenize import RegexpTokenizer
```

```python
# create a tokenizer that uses the regular expressionnto match word char.
removal = RegexpTokenizer(r"\w+")
```

```python
new_text_tokens = [removal.tokenize(i) for i in df["Text"]]
```

```python
new_text_tokens
```

```python
# Import stopwords :
from nltk.corpus import stopwords
```

```python
# create a set of english stopwords :
sw_removal = set(stopwords.words("english"))
```

```python
new_string = [" ".join(i) for i in new_text_tokens]
```

```python
new_string
```

```python
new_str_tokens = [word_tokenize(i) for i in new_string]
```

```python
new_str_tokens
```

```python
# without_stopwords :
without_stopwords = [[word for word in i if not word in sw_removal] for i in new_str_tokens]
```

```python
without_stopwords
```

```python
# Concatenate tokens into a single string for each document
clean_text = [" ".join(i) for i in without_stopwords]
```

```python
clean_text
```

```python
df["Text"] = clean_text
```

```python
df
```

```python
df["Sentiment"].unique()
```

```python
df["Sentiment"] = df["Sentiment"].str.strip()
```

```python
df["Sentiment"].unique()
```

```python
df["Sentiment"].value_counts().head(10)
```

```python
# Import Necesary library :
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
# Fit the tokens :
tokens = Tokenizer()
tokens.fit_on_texts(df["Text"])
```

```python
# Vocaublary size :
vocab_size = len(tokens.word_index)+1
vocab_size
```

```python
seq = tokens.texts_to_sequences(df["Text"])
seq
```

```python
max_length = max([len(i) for i in seq])
max_length
```

```python
pad_seq = sequence.pad_sequences(seq, maxlen=max_length, padding="post")
pad_seq
```

```python
labels = np.array([1 if i.lower() == "positive" else 0 for i in df["Sentimente"]])
labels
```

```python
# Define the model :
model = Sequential([
    Embedding(vocab_size, 20, input_length=max_length),
    SimpleRNN(20),
    Dense(1, activation="sigmoid")
])
```

```python
# Compile the model :
model.compile(optimizer="adam", 
              loss="binary_crossentropy", 
              metrics=["accuracy"])
```

```python
# Train the model :
model.fit(pad_seq, labels, epochs=30, verbose=1)
```
