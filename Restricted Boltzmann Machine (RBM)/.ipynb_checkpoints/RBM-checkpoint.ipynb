```python
# Import library :
from sklearn.neural_network import BernoulliRBM
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
# Create data :
data = make_classification(n_samples=1000, n_classes=3, n_features=10, n_informative=7)
```

```python
data[0]
```

```python
data[1]
```

```python
x = data[0]
y = data[1]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Initialize a StandardScaler :
scaler = StandardScaler()
```

```python
# Fit the StandardScaler to the data in 'x' :
x = scaler.fit_transform(x)
```

```python
# Compute and return the mean :
x.mean()
```

```python
# Compute & return standard deviation :
x.std()
```

- ğ‹ğ¨ğ ğ¢ğ¬ğ­ğ¢ğœ ğ‘ğğ ğ«ğğ¬ğ¬ğ¢ğ¨ğ§ :

```python
# Create the model :
model = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=12)
```

```python
# Train the model :
model.fit(x_train, y_train)
```

```python
# Predict the model :
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import f1_score
```

```python
# print f1-score :
f1_score(y_pred, y_test, average=None)
```
