```python
import pandas as pd
from prophet import Prophet
from prophet.plot import plot, plot_components_plotly, plot_plotly
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("D:\Dataset files_ML\MaunaLoaDailyTemps.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
df = df[["DATE", "AvgTemp"]]
```

```python
df.head()
```

```python
df.columns = ["ds","y"]
```

```python
df.head()
```

```python
df["ds"] = pd.to_datetime(df["ds"])
```

```python
df.head()
```

```python
df.plot(x="ds", y="y", figsize=(15,4))
```

```python
df.tail()
```

```python
len(df)
```

```python
train = df.iloc[:len(df)-365]
test = df.iloc[len(df)-365]
```

```python
model = Prophet()
```

```python
model.fit(train)
```

```python
future = model.make_future_dataframe(periods=365)
```

```python
forecast = model.predict(future)
```

```python
type(forecast)
```

```python
forecast.tail()
```

```python
forecast[["ds","yhat","yhat_lower","yhat_upper"]]
```

```python
train.tail()
```

```python
plt.figure(figsize=(15,4))
plot_plotly(model, forecast)
```

```python
pred = forecast.iloc[-365:]["yhat"]
```

```python
from statsmodels.tools.eval_measures import rmse
```

```python
rmse(pred, test["y"])
```
